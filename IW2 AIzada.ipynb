# Taylor's Polynomial 
### $ f(x) = \sum \limits _{i=0} ^{\infty} \frac{f^{n}(a)}{n!} (x-a)^{n} $  

```python
n = int(input())
c = int (input()) 
Taylors_series = sym.sinh(x)
sym.series(Taylors_series, x, c, n)
```

```python
import math
def sinh_func(x, n):
    sinh_a = 0
    for i in range(n):
        coeff = (-1)**i
        num = x**(2*i)
        denominator = math.factorial(2*i)
        sinh_a += ( coeff ) * ( (num)/(denominator) )
    
    return sinh_a


angles = np.arange(-2*np.pi,2*np.pi,0.1)
sinh_p = np.cos(angles)

fig, ax = plt.subplots()
ax.plot(angles,sinh_p)


for i in range(1,5,2):
    t_sinh = [sinh_func(angle,i) for angle in angles]
    ax.plot(angles,t_sinh)

ax.set_ylim([-7,10])


leg_list = ['sinh function']
for i in range(-8, 9):
    leg_list.append(f'Taylor Series - {i} terms')
ax.legend(leg_list, loc=5)

plt.show()
```

# Maclaurin's Polynomial

```python
x = sym.symbols('x')
n = int(input())
Maclaurin_pol = sym.sinh(x)
Maclaurin_pol.series(x,0, n).removeO()
```

```python
import math

def arccos_func(x, n):
    arccos_a = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i)
        denom = math.factorial(2*i)
        arccos_a += ( coef ) * ( (num)/(denom) )
    
    return arccos_a


angles = np.arange(-2*np.pi,2*np.pi,0.1)
p_arccos = np.arccos(angles)

fig, ax = plt.subplots()
ax.plot(angles,p_arccos)

for i in range(1,6):
    t_arccos = [arccos_func(angle,i) for angle in angles]
    ax.plot(angles,t_arccos)

ax.set_ylim([-7,4])


legend_lst = ['arccos() function']
for i in range(1,6):
    legend_lst.append(f'Maclaurian Series - {i} terms')
ax.legend(legend_lst, loc=8)
    

```

## Approximating pi using arctan1

#### $ pi = 4 * arctan(1)$

#### $ 4(1 - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} ... + \frac{(-1)^{n}}{2n+1})$

#### $ 4 \sum \limits _{n=0} ^{\infty} \frac{(-1)^{n}}{2n+1} $

```python
pi_arctan = 0
for d in range(10000):
    pi_arctan += 4 * (1 - (d % 2) * 2) / (2 * d + 1)
    
print(pi_arctan)
```

# Approximating Pi using Rmanujan's Series 

#### $ 1 = tan(\frac{pi}{4}) $

#### $ 4 arctan \frac{1}{5} - arctan \frac{1}{240} = \frac{pi}{4} $

#### $ 4(4 arctan \frac{1}{5} - arctan \frac{1}{240}) = pi $

```python
from math import atan

def approx_pi():
    return 16 * atan(1/5) - 4 * atan(1/240)


approx_pi()
```
